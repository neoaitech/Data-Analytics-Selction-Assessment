```python
# Business Analysis Questions 
```

```python
1. Which disease contributes the highest total treatment cost?
```

```python
disease_cost = df.groupby("Disease")["Treatment_Cost"].sum().sort_values(ascending=False)
disease_cost
```

```python
2. Which department generates the highest revenue?
```

```python
department_revenue = df.groupby("Department")["Treatment_Cost"].sum().sort_values(ascending=False)
department_revenue
```

```python
3. What is the average treatment cost by disease and insurance status?
```

```python
avg_treatment_cost = (
    df.groupby(["Disease", "Insurance_Status"])["Treatment_Cost"]
      .mean()
      .round(2)
)
avg_treatment_cost
```

```python
4. What percentage of total treatment costs are covered by insured patients?
```

```python
total_cost = df["Treatment_Cost"].sum()

insured_cost = df[df["Insurance_Status"] == "Yes"]["Treatment_Cost"].sum()

insured_percentage = (insured_cost / total_cost) * 100

print(f"Percentage of treatment costs covered by insured patients: {insured_percentage:.2f}%")
```

```python
5. Which city has the highest average treatment cost per patient?
```

```python
city_avg_cost = (
    df.groupby("City")["Treatment_Cost"]
      .mean()
      .sort_values(ascending=False)
      .round(2)
)
city_avg_cost
```

```python
# Patient Demographics Analysis
```

```python
6. Which age group (18–30, 31–45, 46–60, 60+) contributes the highest healthcare expenditure?
```

```python
bins = [18, 30, 45, 60, float('inf')]
labels = ["18-30", "31-45", "46-60", "60+"]

df["Age_Group"] = pd.cut(df["Age"], bins=bins, labels=labels, include_lowest=True)

df["Age_Group"].head()
```

```python
7. What diseases are most common among different age groups?
```

```python
disease_age_group = (
    df.groupby(["Age_Group", "Disease"], observed=False)
      .size()
      .reset_index(name="Patient_Count")
)

disease_age_group
```

```python
8. Is there a significant difference in treatment costs between male and female patients?
```

```python
gender_cost = (
    df.groupby("Gender")["Treatment_Cost"]
      .mean()
      .round(2)
)
gender_cost
```

```python
9. Which blood group has the highest incidence of critical outcomes?
```

```python
critical_patients = df[df["Outcome"] == "Critical"]

blood_group_critical = (
    critical_patients.groupby("Blood_Group")
                     .size()
                     .sort_values(ascending=False)
)
blood_group_critical
```
