```python
df = df.drop_duplicates()
```

```python
df = df.copy()

df['Admission_Date'] = pd.to_datetime(df['Admission_Date'])
df['Discharge_Date'] = pd.to_datetime(df['Discharge_Date'])
```

```python
df.isnull().sum()
```

```python
# 1 Which disease contributes the highest total treatment cost?
```

```python
disease_cost = df.groupby('Disease')['Treatment_Cost'].sum().sort_values(ascending=False)
print(disease_cost)
```

```python
# 2 .Which department generates the highest revenue?
```

```python
dept = df.groupby('Department')['Treatment_Cost'].sum().sort_values(ascending=False)
print(dept)
```

```python
# 3. Average treatment cost by disease and insurance status
```

```python
avg_cost = df.pivot_table(
    values='Treatment_Cost',
    index='Disease',
    columns='Insurance_Status',
    aggfunc='mean'
)
print(avg_cost)
```

```python
# 4.Percentage of total treatment costs covered by insured patients
```

```python
total = df['Treatment_Cost'].sum()

insured = df[df['Insurance_Status']=='Yes']['Treatment_Cost'].sum()

percentage = (insured/total)*100

print(f"Percentage Covered: {percentage:.2f}%")
```

```python
# 5. Which city has the highest average treatment cost?
```

```python
city = df.groupby('City')['Treatment_Cost'].mean().sort_values(ascending=False)
print(city)
```

```python
bins = [18,30,45,60,100]
labels = ['18-30','31-45','46-60','60+']
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=bins,
    labels=labels,
    include_lowest=True
)
df.head()
```

```python
# 6. Which age group contributes the highest healthcare expenditure?
```

```python
age_cost = df.groupby('Age_Group', observed=True)['Treatment_Cost'].sum()
print(age_cost)
```

```python
# 7. Most common disease among different age groups
```

```python
common = (
    df.groupby(['Age_Group', 'Disease'], observed=False)
      .size()
      .reset_index(name='Count')
)
common = common.sort_values(['Age_Group', 'Count'], ascending=[True, False])
print(common)
```

```python
#9. Which blood group has the highest incidence of critical outcomes?
```

```python
critical = df[df['Outcome']=="Critical"]
blood = critical.groupby('Blood_Group').size().sort_values(ascending=False)
print(blood)
```
